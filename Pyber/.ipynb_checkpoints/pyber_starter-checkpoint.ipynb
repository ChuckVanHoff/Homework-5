```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
```

```python
# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
# Combine the data into a single dataset
combined_data_df = ride_data_df.merge(city_data_df, how="left")

# # Display the data table for preview
combined_data_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Get the different types of cities separated
urban_type_df = combined_data_df.loc[(combined_data_df["type"] == "Urban")]
suburban_type_df = combined_data_df.loc[(combined_data_df["type"] == "Suburban")]
rural_type_df = combined_data_df.loc[(combined_data_df["type"] == "Rural")]

# Create a GroupBy city object
city_groupby = combined_data_df.groupby("city")

# collect some aggregate city data:

    # total number of rides per city
ride_counts_df = city_groupby["ride_id"].nunique()
    
    # total number of drivers per city
driver_counts_df = city_data_df["driver_count"]

    # total fares value per city
fare_sums_df = city_groupby["fare"].sum()

    # average fare value per city
average_fare_df = fare_sums_df / ride_counts_df

# Obtain the x and y coordinates for each of the three city types:
    
    # Create a GroupBy city object for each city type 
rural_city_groupby = rural_type_df.groupby("city")
suburban_city_groupby = suburban_type_df.groupby("city")
urban_city_groupby = urban_type_df.groupby("city")

    # total rides per rural, suburban, urban city 
rural_ride_counts_df = rural_city_groupby["ride_id"].nunique()
suburban_ride_counts_df = suburban_city_groupby["ride_id"].nunique()
urban_ride_counts_df = urban_city_groupby["ride_id"].nunique()

    # total drivers per rural, suburban, urban city
rural_driver_counts_df = rural_city_groupby["driver_count"].first()
suburban_driver_counts_df = suburban_city_groupby["driver_count"].first()
urban_driver_counts_df = urban_city_groupby["driver_count"].first()
    
    # total value of fares per rural, suburban, urban city
rural_fare_sums_df = rural_city_groupby["fare"].sum()
suburban_fare_sums_df = suburban_city_groupby["fare"].sum()
urban_fare_sums_df = urban_city_groupby["fare"].sum()

    # average fare value per rural, suburban, urban city
rural_average_fare_df = rural_fare_sums_df / rural_ride_counts_df
suburban_average_fare_df = suburban_fare_sums_df / suburban_ride_counts_df
urban_average_fare_df = urban_fare_sums_df / urban_ride_counts_df

# Set x and y coordinates for rural, suburban, urban cities
rural_x_axis = rural_ride_counts_df
rural_y_axis = rural_average_fare_df

suburban_x_axis = suburban_ride_counts_df
suburban_y_axis = suburban_average_fare_df

urban_x_axis = urban_ride_counts_df
urban_y_axis = urban_average_fare_df

# Build the scatter plots for each city types

plt.scatter(rural_x_axis, rural_y_axis, s=10*rural_driver_counts_df, color="Gold", alpha=.8, edgecolor="k", label="Rural")
plt.scatter(suburban_x_axis, suburban_y_axis, s=10*(suburban_driver_counts_df), color="LightSkyBlue",
           alpha=.8, edgecolor="k", label="Suburban")
plt.scatter(urban_x_axis, urban_y_axis, s=10*urban_driver_counts_df, color="LightCoral",
            alpha=.8, edgecolor="k", label="Urban")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data 2016")
plt.xlabel("Total Numbr of Rides (per City)")
plt.ylabel("Average Fare (per city)")
plt.grid()

# Create a legend
lgnd = plt.legend(frameon=True, edgecolor="lightgray", title="City Types", fontsize=10)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(13,41, "Note: Circle size correlates to \n number of drivers per city.", )

# Save Figure
plt.savefig("RideSharePlot.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
    
    # Get the general total fare value and values from each city type
fares_total = combined_data_df["fare"].sum()
rural_fares_total = rural_type_df["fare"].sum()
suburban_fares_total = suburban_type_df["fare"].sum()
urban_fares_total = urban_type_df["fare"].sum()

# Build Pie Charts
types = ["Rural", "Suburban", "Urban"]
fare_totals = [rural_fares_total, suburban_fares_total, urban_fares_total]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0.0, 0.0, 0.1]


plt.pie(fare_totals, labels=types, colors=colors, autopct="%1.1f%%", explode=explode, shadow=True, startangle=0)

plt.title("Percent of Total Fares per City Type")


# Save Figure
plt.savefig("PyberFareAnalysis.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides = combined_data_df["ride_id"].nunique()
rural_rides_total = rural_type_df["ride_id"].nunique()
suburban_rides_total = suburban_type_df["ride_id"].nunique()
urban_rides_total = urban_type_df["ride_id"].nunique()

# Build Pie Chart
plt.title("Percent of Total Rides by City Type")

types = ["Rural", "Suburban", "Urban"]
driver_counts = [rural_rides_total, suburban_rides_total, urban_rides_total]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0,0,.1]

plt.pie(driver_counts, labels=types, colors=colors, autopct="%1.1f%%", explode=explode, shadow=True, startangle=0)

# Save Figure
plt.savefig("PyberRidesAnalysis.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
    
    # total driver count
total_driver_count = city_data_df["driver_count"].sum()
    
    # total drivers per rural, suburban, urban city
rural_driver_count_total = rural_driver_counts_df.sum()
suburban_driver_count_total = suburban_driver_counts_df.sum()
urban_driver_count_total = urban_driver_counts_df.sum()

    # calculate percents
rural_driver_percent = rural_driver_count_total / total_driver_count
suburban_driver_percent = suburban_driver_count_total / total_driver_count
urban_driver_percent = urban_driver_count_total / total_driver_count

# Build Pie Chart
types = ["Rural", "Suburban", "Urban"]
driver_counts = [rural_driver_count_total, suburban_driver_count_total, urban_driver_count_total]
colors = ["gold", "lightskyblue", "lightcoral"]

plt.title("% of Total Drivers by City Type")

plt.pie(driver_counts, labels=types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)

# Save Figure
plt.savefig("PyberDriverAnalysis.png")
```
